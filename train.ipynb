#Correlation Heatmap

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

print("\nðŸ”Ž Heatmap Observations:")
print("- Survival is negatively correlated with Pclass â†’ higher class = better survival chances.")
print("- Survival is positively correlated with Fare â†’ higher fare = higher chance of survival.")
print("- Age has very weak correlation with survival.")
print("- SibSp and Parch are moderately correlated (family size factor).")
```

#Histogram

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

print("\nðŸ”Ž Age Histogram Observations:")
print("- Most passengers were between 20â€“40 years old.")
print("- Very few children (<10) and elderly (>60).")
```

```python
df['Fare'].hist(bins=30, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

print("\nðŸ”Ž Fare Histogram Observations:")
print("- Most passengers paid a low fare (<100).")
print("- Few passengers paid very high fares (outliers).")
```

#Boxplot

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Survived", y="Age", data=df)
plt.title("Age vs Survival")
plt.show()

print("\nðŸ”Ž Boxplot (Age vs Survival) Observations:")
print("- Survivors had a wide age range, but more younger passengers survived.")
print("- Median age of survivors is slightly lower than non-survivors.")
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare vs Passenger Class")
plt.show()

print("\nðŸ”Ž Boxplot (Fare vs Class) Observations:")
print("- Higher class passengers (1st class) paid much higher fares.")
print("- 3rd class passengers mostly paid very low fares.")
```

#Scatterplot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare (Survival)")
plt.show()

print("\nðŸ”Ž Scatterplot (Age vs Fare with Survival) Observations:")
print("- Survivors are more concentrated among higher fare passengers.")
print("- Younger passengers also have slightly better survival chances.")
print("- Many low-fare 3rd class passengers did not survive.")
```

#Final Summary

```python
print("\nðŸ“Š Summary of Findings:")
print("1. 1st class passengers had higher survival rates, while 3rd class had the lowest.")
print("2. Higher fare strongly increased survival chances (wealthier passengers benefited).")
print("3. Age had weak correlation; children survived slightly more often, but overall effect was small.")
print("4. Family size mattered â€” small families had better survival than large families.")
print("5. Most passengers were young adults (20â€“40) and paid low fares; many of them in 3rd class did not survive.")
```
